# 模型选择、欠拟合、过拟合

***通过多项式拟合来交互地探索这些概念***

```python
import math
import numpy as np
import torch
from torch import nn
from d2l import torch as d2l
```


最后那个是噪音

```python
#多项式的最大阶数，确定w向量的维度（1，20)
max_degree=20
n_train,n_test=100,100 #训练样本和测试样本（验证样本）
true_w=np.zeros(max_degree) #真是权重和输入特征一样，都是20，分配大量空间
true_w[0:4]=np.array([5,1.2,-3.4,5.6])#初始化权重，前四个是确定值，后面16个全是0

#生成（200，1）的特征
features = np.random.normal(size=(n_train + n_test, 1))
#将生成的特征打乱
np.random.shuffle(features)
#数组元素求n次方 生成（200，20)的向量 reshape(1,-1)转化为1行，列数自动计算
poly_features = np.power(features, np.arange(max_degree).reshape(1, -1))
for i in range(max_degree):
    poly_features[:, i] /= math.gamma(i + 1) #math.gamma()求阶乘
#权重和后面部分相乘 label的维度（n_train+n_test)
labels = np.dot(poly_features, true_w)
#scale表示正态分布的标准差
labels += np.random.normal(scale=0.1, size=labels.shape) #加入噪音
```

```python
true_w, features, poly_features, labels = [torch.tensor(x, dtype=torch.float32) #numpy转换为tensor
    for x in [true_w, features, poly_features, labels]]

#观察具体特征类型
features[:2], poly_features[:2, :], labels[:2]
```

```python
#观察每个变量的具体形状
true_w.shape, features.shape, poly_features.shape, labels.shape
```

```python
def evaluate_loss(net, data_iter, loss):  
    """评估给定数据集上模型的损失。"""
    #d2l.Accumulator(2)创建2个单位，存放总loss和单位数量
    metric = d2l.Accumulator(2) #损失的总和，样本数量
    #传入数据data_iter
    for X, y in data_iter:
        #out为预测y的值
        out = net(X)
        #y为真实的值
        y = y.reshape(out.shape)
        #根据定义的loss计算损失函数
        l = loss(out, y)
        #l.sum()总损失，l.numel()相当于总共l的数量
        metric.add(l.sum(), l.numel())
    return metric[0] / metric[1]
```

```python
#传入训练特征，测试特征，训练标签，测试标签，迭代次数
def train(train_features, test_features, train_labels, test_labels,num_epochs=400):
    loss = nn.MSELoss()
    #输入形状为train_features的列数，从后面开始算，也就是20
    input_shape = train_features.shape[-1]
    #用了一个线性层，不设置偏置，因为已经在多项式中实现了
    #input_shape是输入维度，1是输出维度
    net = nn.Sequential(nn.Linear(input_shape, 1, bias=False))
    #选择小批量大小
    batch_size = min(10, train_labels.shape[0])
    #train_labels.reshape(-1, 1)是将标签变成1列
    #d2l.load_array传入数据
    train_iter = d2l.load_array((train_features, train_labels.reshape(-1, 1)),batch_size)
    test_iter = d2l.load_array((test_features, test_labels.reshape(-1, 1)),batch_size, is_train=False)
    #选择SGD去更新w和b
    trainer = torch.optim.SGD(net.parameters(), lr=0.01)
    #画图 动态显示训练和验证集损失的情况
    animator = d2l.Animator(xlabel='epoch', ylabel='loss', yscale='log',xlim=[1, num_epochs], ylim=[1e-3, 1e2], legend=['train', 'test'])
    #num_epochs=400
    for epoch in range(num_epochs):
        #d2l.train_epoch_ch3表示迭代一次
        d2l.train_epoch_ch3(net, train_iter, loss, trainer)
        #每20次画一次图
        if epoch == 0 or (epoch + 1) % 20 == 0:
            animator.add(epoch + 1, (evaluate_loss(net, train_iter, loss), evaluate_loss(net, test_iter, loss)))
    #默认w的输出
    print('weight:', net[0].weight.data.numpy())
```

```python
# 从多项式特征中选择前4个维度，即1,x,x^2/2!,x^3/3! 输出的权重和多项式中的差不多
#n_train=100,一般训练集一半测试集
train(poly_features[:n_train, :4], poly_features[n_train:, :4],labels[:n_train], labels[n_train:])
```

```python
#取前两个维度训练 欠拟合
train(poly_features[:n_train, :2], poly_features[n_train:, :2],labels[:n_train], labels[n_train:])
```

```python
#取所有维度进行训练 过拟合
train(poly_features[:n_train, :], poly_features[n_train:, :],labels[:n_train], labels[n_train:], num_epochs=1500)
```

## 答疑

+ 神经网络本身是一种语言  
+ 严格来说，泛化误差是指那些不在手上的数据的泛化误差
+ 训练数据集上作K折交叉验证，数据集不够大的时候才用，深度学习一般不做，因为比较昂贵
+ 设计超参数，一个是靠经验，二是搜索，网格是全部试，可能超参数空间很大，不能做，随机比较推荐，贝叶斯的话，也是要训练100次，1000次，hpo  
+ k折交叉验证目的是确定超参数，接下来有三种做法，一是再用这个超参数训练一遍数据，二是选择k折交叉验证过程中最好的那个作为最后的结果，这样训练样本会少一些，精度会损失一些，三是将测试数据集放到k个模型中都预测一次，取平均，这样测试成本会增加，但是因为取均值，稳定性好  
+ 随机森林很少有做进深度学习的，主要是因为梯度不好传，不能用梯度下降
+ 课程里面的标号就是label  
+ CNN本质上就是MLP，只是限制了一些权重  
+ 同样的模型，同样的数据，只是随机初始化不一样，模型都会有点偏移，做n次可以降低一些放差，可能每次模型走的路都不一样，好像把你丢在大山里，随便走一走，走出来的路都不一样，一般来说集成会好一点
+ 数据集噪声当然是希望越少越好
+ 加权 权重 样本不平衡 1：9，先看是数据集本来就应该是这样的比例，还是因为采样没采样好，若是采样没采样好，一个是在损失函数里给小比例一个更大的权重，一个是把1复制9次，这样样本就平衡了。也就是说，在深度学习里，发生问题先考虑逻辑上应该怎么做，而不是被代码唬住
+ 验证数据集先下降再上升这就是过拟合

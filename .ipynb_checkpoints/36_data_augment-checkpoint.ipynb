## 自定义数据增强

```python
#num_rows=2,num_cols=4是指增强后的图片以2行4列排，相当于把aug作用在img上8次，scale=1.5是指画出来的尺寸大小
def apply(img,aug,num_rows=2,num_cols=4,scale=1.5):
    for _ in range(num_rows*m_cols):
        aug(img)
        Y=[aug]
    d2l.show_images(Y,num_rows,num_cols,scale=scale)
```
